## Script of Stopwords

```python
stops = pd.read_csv(stopwords, header=None)[0].values.tolist()
```

```python
def remove_stopwords(text):
    tokens = text.lower().split()
    cleaned_tokens = [token for token in tokens if token not in stops]
    return (" ".join(cleaned_tokens))
```

## Script of Special Chars

```python
# sometimes, query containing special chars like /,', or ? will produce errors
# https://github.com/terrier-org/pyterrier/issues/62
def remove_specials(text):
  pattern = re.compile('[\W_]+')
  return pattern.sub(' ', text)
```

## Script of Normalization

```python
df_kamus_baku = pd.read_csv(kamus_baku, index_col=0)

def normalisasi_singkatan(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    token_after_normalized = []

    for token in tokens:
        if token in df_kamus_baku['singkatan'].values:
            token = df_kamus_baku.loc[df_kamus_baku['singkatan'] == token, 'asli'].values[0]
        token_after_normalized.append(token)

    normal_text = ' '.join(token_after_normalized)
    return normal_text

def normalisasi_kata_ulang(text):
    text = text.lower()
    # contoh: kira2 -> kira-kira
    for found in re.finditer(r'[A-z]+(2|2x)\b', text):
        found = found.group(0)
        replace = re.sub(r'(2|2x)', "", found)
        replace = replace + '-' + replace
        text = text.replace(found, replace)

    # contoh: apa apa -> apa-apa
    for found in re.finditer(r'\b([A-z]+)\b\s+\1', text):
        found = found.group(0)
        word = found.split(' ');
        word = word[0] + "-" + word[0]
        text = text.replace(found, word)
    return text

def normalisasi(question):
    question = normalisasi_singkatan(question)
    question = normalisasi_kata_ulang(question)

    return question
```

```python
%%time

xtree = et.parse("full_dataset.xml")
xroot = xtree.getroot()
rows = []

for node in xroot:
  id = node.find("ID").text.strip() if node.find("ID").text is not None else None
  url = node.find("URL").text.strip() if node.find("URL").text is not None else None
  tag = node.find("TAG").find("ITEM").text.strip() if node.find("TAG").text is not None else None
  keluhan = node.find("KELUHAN")
  keluhan_judul = keluhan.find("JUDUL").text.strip() if keluhan.find("JUDUL").text is not None else None
  keluhan_isi = keluhan.find("ISI").text.strip() if keluhan.find("ISI").text is not None else None
  keluhan_waktu = keluhan.find("WAKTU").text.strip() if keluhan.find("WAKTU").text is not None else None
  jawaban = node.find("JAWABAN")
  jawaban_isi = jawaban.find("ISI").text.strip() if jawaban.find("ISI").text is not None else None
  jawaban_waktu = jawaban.find("WAKTU").text.strip() if jawaban.find("WAKTU").text is not None else None

  rows.append({"docno": id, "url": url, "tag": tag, \
                "keluhan_judul": keluhan_judul, "keluhan_isi": keluhan_isi, "keluhan_waktu": keluhan_waktu, \
                "jawaban_isi": jawaban_isi, "jawaban_waktu": jawaban_waktu})

data = pd.DataFrame(rows, columns = ['docno', 'url', 'tag', 'keluhan_judul', 'keluhan_isi', 'keluhan_waktu', 'jawaban_isi', 'jawaban_waktu'])
```

```python
# feature adalah fungsi dari pasangan dokumen
# dan query --> feature(dokumen, query)
def avg_vector_embedding(model_embedding, tokenized_data):
  """
  Representasi sederhana dokumen sebagai rata-rata vektor
  hasil word embedding dari seluruh kata yang ada di dokumen tersebut
  """
  embedding = []
  vector_acc = np.zeros(VECTOR_SIZE)
  count = 0
  for word in tokenized_data:
    if word in model_embedding.wv:
      vector_acc += model_embedding.wv[word]
      count += 1
  if count > 0:
    vector_acc /= count
  embedding.append(vector_acc)
  return np.array(embedding)

def f_len_diff(text, query):
  """ Delta panjang dokumen & query """
  return abs(len(text.split()) - len(query.split()))

def f_jaccard(text, query):
  """ Jaccard Similarity """
  word_set_text = set(text)
  word_set_query = set(query)
  return len(word_set_text & word_set_query) / len(word_set_text | word_set_query)

def f_canberra(text_embedding, query_embedding):
  """ Canberra Distance """
  return distance.canberra(text_embedding, query_embedding)

def f_cosine(text_embedding, query_embedding):
  """ Cosine Similarity """
  return cosine(text_embedding, query_embedding)

def f_wmd(model_embedding, text, query):
  """ Word Mover Distance """
  word_text = text.split()
  word_query = query.split()

  MAX_DISTANCE_VALUE = 10.0

  if all(token in model_embedding.wv for token in word_text) and all(token in model_embedding.wv for token in word_query):
      distance = model_embedding.wv.wmdistance(word_text, word_query)
  else:
      distance = MAX_DISTANCE_VALUE

  return distance

def f_rake_jaccard(text, query):
  """ Jaccard Distance with RAKE """
  r_text = Rake(stops)
  r_text.extract_keywords_from_text(text)

  r_query = Rake(stops)
  r_query.extract_keywords_from_text(query)

  return f_jaccard(r_text.get_ranked_phrases(), r_query.get_ranked_phrases())

def f_common_distinct_words(text, query):
  word_set_text = set(text.split())
  word_set_query = set(query.split())
  return len(word_set_text & word_set_query)

def combine_f(docno, qid, text, query):
  # text_w2v_embedding = avg_vector_embedding(model_w2v, text.split())
  # query_w2v_embedding = avg_vector_embedding(model_w2v, query.split())
  text_ft_embedding = avg_vector_embedding(model_ft, text.split())
  query_ft_embedding = avg_vector_embedding(model_ft, query.split())
  return np.array([
                    f_len_diff(text, query), \
                    # f_jaccard(text.split(), query.split()), \

                    # f_cosine(bert_doc_embeddings[docno], bert_query_embeddings[qid]), \
                    # f_cosine(t5_doc_embeddings[docno], t5_query_embeddings[qid]), \
                    # f_cosine(pminilm_doc_embeddings[docno], pminilm_query_embeddings[qid]), \
                    # f_cosine(text_w2v_embedding.flatten(), query_w2v_embedding.flatten()), \
                    f_cosine(text_ft_embedding.flatten(), query_ft_embedding.flatten()), \
                    # f_cosine(preprocessing.normalize(text_w2v_embedding).flatten(), preprocessing.normalize(query_w2v_embedding).flatten()), \
                    # f_cosine(preprocessing.normalize(text_ft_embedding).flatten(), preprocessing.normalize(query_ft_embedding).flatten()), \

                    # f_wmd(model_w2v, text, query), \
                    f_wmd(model_ft, text, query), \

                    # f_canberra(text_w2v_embedding.flatten(), query_w2v_embedding.flatten()), \
                    # f_canberra(preprocessing.normalize(text_w2v_embedding).flatten(), preprocessing.normalize(query_w2v_embedding).flatten()), \
                    # f_canberra(text_ft_embedding.flatten(), query_ft_embedding.flatten()), \

                    f_common_distinct_words(text, query), \

                    # f_rake_jaccard(text, query)
                   ])
```

```python
%%time

""" Model word embedding """
tokenized_corpus = []
for doc in data['keluhan']:
  tokenized_corpus.append(doc.split())
VECTOR_SIZE = 64 # masih arbitrary saja
model_w2v = Word2Vec(tokenized_corpus, vector_size=VECTOR_SIZE, min_count=2, workers=1) # CBOW
model_ft = FastText(tokenized_corpus, vector_size=VECTOR_SIZE, min_count=2, workers=1)

""" KFold """
FOLD = 5
kf = KFold(n_splits = FOLD, shuffle=True, random_state=42)

evaluation_results = pd.DataFrame()
metrics = dict()
metric_names = [P@10, P@5, "map", "recip_rank", nDCG@5]
model_names = ["BM25 Only", "BM25 >> LambdaMART XGBoost"]

""" Cross validation """
for train_idx, test_idx in kf.split(queries):
  train_queries, test_queries = queries.iloc[train_idx], queries.iloc[test_idx]

  more_features = pt.apply.doc_features(lambda row: combine_f(row["docno"], row["qid"], row["text"], row["query"]))
  pipeline = (bm25 % CUT_OFF) >> pt.text.get_text(index_ref, "text") >> (more_features ** (bm25 ** bb2))
  # pipeline = (bm25 % CUT_OFF) >> pt.text.get_text(index_ref, "text") >> (more_features ** (bm25 ** bb2 ** cross_encT))
  # pipeline = (bm25) >> pt.text.get_text(index_ref, "text") >> (more_features ** (bm25 ** bb2 ** cross_encT))

  lmart_xgb = xgb.sklearn.XGBRanker(objective = 'rank:ndcg',
                              learning_rate = 0.1,
                              gamma = 1.0,
                              min_child_weight = 0.1,
                              max_depth = 6,
                              random_state = 42)

  lmart_xgb_pipeline = pipeline >> pt.ltr.apply_learned_model(lmart_xgb, form = "ltr")
  lmart_xgb_pipeline.fit(train_queries, qrels, test_queries, qrels)

  evaluation_results = pd.concat([evaluation_results, \
                                  pt.Experiment([(bm25 % CUT_OFF), lmart_xgb_pipeline], \
                                                test_queries, \
                                                qrels, \
                                                eval_metrics=metric_names, \
                                                names=model_names, \
                                                perquery=True)], \
                                  ignore_index=True)

""" Konversi """
for index, line in evaluation_results.iterrows():
  model_name = line['name']
  qid = line['qid']
  measure = line['measure']
  score = line['value']

  if measure in metrics:
    if model_name in metrics[measure]:
      metrics[measure][model_name][qid] = score
    else:
      metrics[measure][model_name] = {qid: score}
  else:
    metrics[measure] = {model_name: {qid: score}}
```

```python
def get_serp_metadata(serp):
  return pd.merge(serp, qrels[['qid', 'docno', 'doc_raw', 'label']], how='left', on=['qid', 'docno'])

def get_relevant_serp(serp):
  relevant_docno = pd.merge(serp, qrels[(qrels['label'] == 1)], how='inner', on='docno')['docno'].unique()
  return serp[serp['docno'].isin(relevant_docno)]

def compare_rank_score(metric_scores):
  df_scores = pd.DataFrame(metric_scores)
  print("Rerank < Baseline (Baseline > Rerank)")
  display(df_scores[df_scores['BM25 >> LambdaMART XGBoost'] < df_scores['BM25 Only']])
  print()
  print("Rerank > Baseline (Baseline < Rerank)")
  display(df_scores[df_scores['BM25 >> LambdaMART XGBoost'] > df_scores['BM25 Only']])

def display_observed_query(observed_query):
  print(f"{observed_query}: {queries[queries['qid'] == observed_query]['query'].iloc[0]}")

def display_serp(observed_query, model, get_qrels_label=True, k=10):
  model_result = model.transform(queries[queries['qid'] == observed_query])
  if get_qrels_label:
    model_result = get_serp_metadata(model_result)
  top_metadata = model_result[model_result['rank'] < k].sort_values(by=['rank'])
  display(top_metadata)
```

# Sanity Check (Manual Evaluation)

```python
def prec(serp, k):
  # (berapa yg benar dari top k) / (k)
  return len(get_relevant_serp(serp[:k])) / k

def recip_rank(serp):
  return 1 / (get_relevant_serp(serp).iloc[0]['rank']+1)

def get_dcg(serp, k):
  score = 0.
  for i in range(1, k+1):
    score += (serp.iloc[i-1]['label'])/(math.log2(i+1))
  return score

def ndcg(serp, k):
  serp_metadata = get_serp_metadata(serp)
  dcg = get_dcg(serp_metadata, k)
  idcg = get_dcg(serp_metadata.sort_values(by=['label'], ascending=False), k)
  return dcg/idcg

def sanity_check(serp):
  print(f"P@10 = {prec(serp, 10)}")
  print(f"recip_rank = {recip_rank(serp)}")
  print(f"ndcg@5 = {ndcg(serp, 5)}")

  display(get_relevant_serp(serp))
```
